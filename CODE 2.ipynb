```python
cap = cv.VideoCapture("naruto.mp4")

for frame_index in range(int(cap.get(cv.CAP_PROP_FRAME_COUNT))):    
    ret, frame = cap.read()

    cv.imshow("Video Player", frame)

    if cv.waitKey(0) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

**To play at original speed**

```python
import cv2 as cv

cap = cv.VideoCapture("naruto.mp4")
fps = cap.get(cv.CAP_PROP_FPS)

while True:    
    ret, frame = cap.read()

    cv.imshow("Video Player", frame)

    if cv.waitKey(int(1000/fps)) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
